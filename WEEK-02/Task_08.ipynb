```python
import pandas as pd

# Series
s = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
print(s)

# DataFrame
data = {'Name': ['yasir', 'Uneeba', 'Umer','Momina'],
        'Age': [10, 20, 23,18],
        'City': ['Lahore', 'Karachi', 'Islamabad','Rawalpindi']}
df = pd.DataFrame(data)

#  Table Styling
df.style.set_table_styles(
    [{'selector': 'th',
      'props': [('background-color', '#405067'),
                ('color', 'white'),
                ('font-weight', 'bold')]}]
).highlight_max(color='lightblue').highlight_min(color='plum')
```

```python
# 5. Data Cleaning
df.fillna(0, inplace=True)
df.dropna(inplace=True)
```

```python
# 6. Indexing & Selection
print("3rd index of array (Name):", df.loc[0, "Name"])
print("1st index of array (Age):", df.iloc[1, 2])
```

```python
# 7. Sorting & Filtering
filtered = df[df["Age"] > 23]
print("Filtered Data that is graeater than age (23):", filtered)
sorted_df = df.sort_values(by="Age", ascending=False)
print("The sorted data set by maximum age is:", sorted_df)
```

```python
# 8. GroupBy & Aggregations
grouped = df.groupby("City")["Age"].mean()
print("The data combined on the basis of city and (ages by their mean values):", grouped)
```

```python
# 9. Merging & Joining
data1 = pd.DataFrame({"ID": [1,2,3,4], "Name": ["'Uneeba", "Umer", "yasir","Momina"]})
data2 = pd.DataFrame({"ID": [1,2,3,4], "Salary": [50000, 60000, 55000,0]})
merged = pd.merge(data1, data2, on="ID")
df.style.background_gradient(cmap="Blues")
print("The merged data set by ID (Name) is :", merged)
concat_df = pd.concat([data1, data2], axis=1)
print("The concatenated dataset by ID (Sallery) is :", concat_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
penguins = sns.load_dataset("penguins")

# labels data:
x = penguins["bill_length_mm"]
y = penguins["bill_depth_mm"]

# Subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Line Plot
axs[0,0].plot(penguins["flipper_length_mm"].head(20))
axs[0,0].set_title("Line Plot")

# Bar Chart
penguins["species"].value_counts().plot(kind="bar", ax=axs[0,1], color="purple")
axs[0,1].set_title("Bar Chart")

# Scatter Plot
axs[1,0].scatter(x, y, color="red")
axs[1,0].set_title("Scatter Plot")

# Box Plot
sns.boxplot(x="species", y="body_mass_g", data=penguins, ax=axs[1,1])
axs[1,1].set_title("Box Plot")

plt.tight_layout()
plt.show()
```

```python
# Seaborn styling
sns.set_style("whitegrid")
sns.set_palette("muted")

# Histogram of flipper length
sns.histplot(penguins["flipper_length_mm"].dropna(), kde=True)
plt.title("Flipper Length Distribution (Penguins)")
plt.show()
```

```python

# Violin Plot: species vs bill length
sns.violinplot(
    x="species", 
    y="bill_length_mm", 
    hue="species",         
    data=penguins, 
    palette="pastel", 
    legend=False     
)

plt.title("Violin Plot: Bill Length by Species")
plt.show()
```

```python
# 20. Hybrid Visualization
sns.histplot(penguins["flipper_length_mm"].dropna(), 
             kde=False, color="lightblue", bins=15, label="Histogram")

sns.kdeplot(penguins["flipper_length_mm"].dropna(), 
            color="red", label="KDE")

plt.title("Hybrid Visualization: Flipper Length Distribution")
plt.legend()
plt.show()
```
